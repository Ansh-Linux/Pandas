```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [25, 30, 35, 40],
    "City": ["New York", "Los Angeles", "Chicago", "Houston"],
    "salary": [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)
print(df)

#Now assigning custom indeices to the DataFrame
df.index = ["A", "B", "C", "D"]
print(df)
```

## Reading from a CSV File

```python
df = pd.read_csv("data.csv")

df.head()
df.tail()

#Using different reading methods

df = pd.read_csv('data.csv', sep=";") # If the data is separated by semicolons
print(df)
df = pd.read_csv('data.csv', header=None) # If the data does not have a header row
print(df)
df = pd.read_csv('data.csv', names=['Column1', 'Column2', 'Column3']) # If you want to assign custom column names
print(df)
df = pd.read_csv('data.csv', nrows=2) # To read only the first 2 rows of the data
print(df)
```

## Essentials after loading a dataset using pandas

```python
df = pd.read_csv("data.csv")

df.head() #Checking if the data is loaded correctly
df.dtypes #Checking the data types of each column
df.shape #Checking how big the dataFrame is
df.isnull().sum() #Checking for missing values
df.describe() #Getting a statistical summary of the data
df.info() #Getting a concise summary of the data
```
