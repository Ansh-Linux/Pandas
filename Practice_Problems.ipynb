```python
import pandas as pd 

df = pd.DataFrame({
    "name" : ["Alice", "Bob", "Charlie", "David"],
    'scores' : [85, 92, 78, 90],
    "subject" : ["Math", "Science", "History", "English"]
})

print(df.shape)
print(df.columns)
df.head(2)
```

```python
high_scores = df[df["scores"]>70]
result = high_scores[["name","scores"]]
print(result)
```

```python
grp_subjects = df.groupby("subject")["scores"].mean()
print(grp_subjects)

best = grp_subjects.max()
print(best)
```
