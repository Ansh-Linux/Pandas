```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name":          ["Ali","Sara","Raj","Mia","Tom","Zoe"],
    "math_score":    [88, 72, 95, 61, 84, 77],
    "english_score": [75, 90, 68, 85, 79, 92],
    "age":           [20, 21, 20, 22, 21, 20]
})

mean_maths = np.mean([df["math_score"]])
std_maths = np.std([df["math_score"]])
print(mean_maths, std_maths)
```

```python
df["total_scores"] = df['math_score'] + df['english_score']

df["grade"] = df['total_scores'].apply(lambda x: 'A' if x>= 160 else 'B')
print(df['grade'])
```

```python
scores = df[["math_score","english_score"]].values

normalised = (scores - scores.min(axis=0)) / (scores.max(axis=0) - scores.min(axis=0))

df["math_norm"] = normalised[:, 0]
df["english_norm"] = normalised[:, 1]

print(df[["english_norm", "math_norm"]])
```

```python
df2 = pd.DataFrame({
    "name":          ["Ali","Sara","Raj","Mia","Tom"],
    "math_score":    [88, np.nan, 95, np.nan, 84],
    "english_score": [75, 90, np.nan, 85, 79]
})

print(df2.isnull().sum())

df2["math_score"] = df2["math_score"].fillna(df2["math_score"].mean())

df2 = df2.dropna(subset=["english_score"])
print(df2)
```

```python
df = pd.DataFrame({
    "name":          ["Ali","Sara","Raj","Mia","Tom","Zoe"],
    "math_score":    [88, 72, 95, 61, 84, 77],
    "english_score": [75, 90, 68, 85, 79, 92],
    "age":           [20, 21, 20, 22, 21, 20]
})

mean_math = np.mean(df["math_score"])
std_math = np.std(df["math_score"])

print(mean_math, std_math)
```

```python
scores = df[["math_score", "english_score"]].values
print(scores)

normalized = (scores - scores.min(axis=0)) / (scores.max(axis=0) - scores.min(axis=0))

df["math_norm"] = normalized[:, 0]
df["english_norm"] = normalized[:, 1]

print(df[["english_norm", "math_norm"]])
```

```python
df["total_score"] = df["math_score"] + df["english_score"]

df["grade"] = df["total_score"].apply(lambda x: "A" if x>=160  else "B")
print(df[["name", "total_score", "grade"]])
```

```python
df2 = pd.DataFrame({
    "name":          ["Ali","Sara","Raj","Mia","Tom"],
    "math_score":    [88, np.nan, 95, np.nan, 84],
    "english_score": [75, 90, np.nan, 85, 79]
})

print(df2.isnull().count())

df2["math_score"] = df2["math_score"].fillna(df2["math_score"].mean())
print(df2["math_score"])

df2 = df2.dropna(subset=["math_score", "english_score"])
print(df2)
```
